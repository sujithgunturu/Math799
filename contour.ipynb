```
import numpy as np
import pandas as pd
df = pd.read_excel('2167869.xlsx')
df.head()
```

```
Z = df.RMSE.values.tolist()
X = df.LATITUDE.values.tolist()
Y = df.LONGITUDE.values.tolist()
```

```
from scipy.interpolate import interp1d
import plotly.express as px
```

```
f = interp1d([0, max(Z)], [5, 18])
circle_radius = f(Z)
```

```
typelist = ['open-street-map']
fig = px.density_mapbox(df, lat = 'LATITUDE', lon ='LONGITUDE', radius = circle_radius, zoom = 0, mapbox_style = 'open-street-map')
fig.show()
```

```
fig = px.density_contour(df, x =X, y= Y,z =Z)
fig.show()
```

```
import plotly.graph_objects as go
fig = go.Figure(go.Densitymapbox(lat=df.LATITUDE, lon=df.LONGITUDE, z=df.RMSE,
                                 radius=10))
fig.update_layout(mapbox_style="stamen-terrain", mapbox_center_lon=180)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
